```python
# V2 create n lag features for feature
def lag_features(df, feature, number_lags):
    for lag in xrange(1, number_lags + 1):
        df['lag_' + str(lag) + '_' + feature] = 0
    target = df.loc[df['ncodpers'] == cust][feature]
    for cust in list(set(df['ncodpers'])):
        for lag in xrange(1, number_lags + 1):
            df.loc[(df['ncodpers'] == cust), 'lag_' + str(lag) + '_' + feature] = target.shift(lag)
```
