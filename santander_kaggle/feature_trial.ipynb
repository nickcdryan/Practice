```python
# create n lag features for feature
def lag_features(df, feature, number_lags):
    for lag in xrange(1, number_lags + 1):
        df['lag_' + str(lag) + '_' + feature] = "Null"
    for cust in unique_customers:
        for lag in xrange(1, number_lags + 1):
            df.loc[(df['member'] == cust), 'lag_' + str(lag) + '_' + feature] = df.loc[df['member'] == cust][feature].shift(lag)
```

```python
# log date and amount
def status_change(df, feature):
    change_log = feature + "_change"
    amount_log = feature + "_amount"
    df[change_log] = 0
    df[amount_log] = 0
    for cust in unique_customers:
        for date in unique_dates[1:]:
            member = df.loc[df['member'] == cust]
            current  = member.loc[member['date'] == date]
            previous = member.loc[member['date'] == (date-1)]
            change = current[feature] !=  previous[feature]
            try:
                amount = float(current[feature]) - float(previous[feature])
            except ValueError:
                amount = "N/A"
            if change.any() == True:
                df.loc[(df["member"] == cust), change_log] = date
                df.loc[(df["member"] == cust), amount_log] = amount
            else:
                continue
```

```python
status_change(df, 'income')
status_change(df, 'married')
status_change(df, 'product5')
status_change(df, 'address')

df
```
