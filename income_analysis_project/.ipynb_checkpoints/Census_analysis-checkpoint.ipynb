#### Plotting the data

Now we'll create some quick univariate feature plots. In defining our plot functions we want to:
- 1) Separate out the features with lots of distinct values so that we can bin values and adjust axes/labels appropriately. We also want to visualize these features twice: once with 0 included (remember, 0 makes up a huge proportion of these features) and once with 0 excluded.   
- 2) Sort nominal feature responses in descending order of number of responses, left to right AND sort something continous like "age" by value. Both conditions make the plots a little more interpretable. We'll also want to differentiate between continuous variables like "age" against integer-type variables like "industry recode," which are nominal/ordinal and do not lend themselves to the same degree of statistical analysis.  
- 3) Visualize features once with null value included, once without null values.

```python
# functions to split and visualize our data

# filter out features with more than 99 unique values
def get_small_feats(df):
    feats = list(df)
    plottable_feats = []
    for i in feats:
        if len(df[i].unique()) < 100:
            plottable_feats.append(i)
    return plottable_feats


# get features with more than 99 unique values
def get_large_feats(df):
    feats = list(df)
    plottable_feats = []
    for i in feats:
        if len(df[i].unique()) >= 100:
            plottable_feats.append(i)
    return plottable_feats
  
    
# split into string-value features, continuous features, and nominal type.int features
# bar plot and include summary statistics where appropriate     
def plot_small_feats(df):
    for feature in get_small_feats(df):
        if df[feature].dtype == 'O':
            df[feature].value_counts().plot.bar(figsize = (15,15))
            plt.title(feature, fontsize = 20)
            plt.show()  
            tempdf = pd.DataFrame(df[feature].value_counts())
            tempdf.loc[:, 'percentage'] = (tempdf.loc[:, feature] / tempdf.loc[:, feature].sum()) * 100
            print tempdf   
        elif (feature == 'age') or (feature == 'weeks worked in year') or (feature == 'num persons worked for employer'):
            df[feature].value_counts(ascending=False, sort=False).plot.bar(figsize = (15,15))
            plt.title(feature, fontsize = 20)
            plt.show()
            print df[feature].describe()            
        else:
            df[feature].value_counts(ascending=False, sort=False).plot.bar(figsize = (15,15))
            plt.title(feature, fontsize = 20)
            plt.show()  
            tempdf = pd.DataFrame(df[feature].value_counts())
            tempdf.loc[:, 'percentage'] = (tempdf.loc[:, feature] / tempdf.loc[:, feature].sum()) * 100
            print tempdf  
            
            
# remove null values from string-value features, keep 0 in continuous cases, remove 0 from nominal type.int features
# include summary statistics where appropriate            
def plot_small_feats_no_null(df):
    for feature in get_small_feats(df):
        if df[feature].dtype == 'O':
            exclude = ['Not in universe', 'Not in universe or children', 'All other', '?', 
                       'Not in universe under 1 year old']
            filtered = df[feature].map(lambda val: not any(i in val for i in exclude))
            df[filtered][feature].value_counts().plot.bar(figsize=(15,15))
            plt.title(feature, fontsize = 20)
            plt.show()  
            tempdf = pd.DataFrame(df[feature].value_counts())
            tempdf.loc[:, 'percentage'] = (tempdf.loc[:, feature] / tempdf.loc[:, feature].sum()) * 100
            print tempdf
        elif (feature == 'age') or (feature == 'weeks worked in year') or (feature == 'num persons worked for employer'):
            df[feature].value_counts(ascending=False, sort=False).plot.bar(figsize = (15,15))
            plt.title(feature, fontsize = 20)
            plt.show()
            print df[feature].describe()
        else:
            filtered = df[df[feature]!=0]
            filtered[feature].value_counts().plot.bar(figsize = (15,15))
            plt.title(feature, fontsize = 20)
            plt.show()
            tempdf = pd.DataFrame(df[feature].value_counts())
            tempdf.loc[:, 'percentage'] = (tempdf.loc[:, feature] / tempdf.loc[:, feature].sum()) * 100            
            print tempdf
            
            
# plot nominal features by variable size (e.g. education), plot continuous features by variable value (e.g. age), though we don't need it for this dataset
# pie plot for large features (includes "0" value), then binned histogram (excludes "0" value)            
def plot_large_feats(df):
    for feature in get_large_feats(df):
        if df[feature].dtype == 'O':
            df[feature].value_counts().plot.pie(figsize = (15,15))
            plt.title(feature, fontsize = 20)
            plt.show()  
        else:
            df[feature].value_counts(ascending=False, sort=False).plot.pie(figsize = (15,15))
            plt.title(feature, fontsize = 20)
            plt.show() 
        
        
# plot large features but exclude value = 0. include summary statistics
# binned histogram for large features
def plot_large_feats_no_0(df):
    for feature in get_large_feats(df):
            filtered = df[df[feature]!=0]
            filtered[feature].plot.hist(figsize = (15,15), bins = 100)
            plt.title(feature, fontsize = 20)
            plt.show()  
            print filtered[feature].describe()        
        
```

#### Machine learning with categorical features: what to do?

Next we'll have to encode non-integer type values as integers. This is because sklearn classifiers need float values for fit and transform functions. 

To do this we have a few options. We can manually encode categorical features like we did for income, mapping our values to 0 and 1. We can also use a method like sklearn's LabelEncoder, which uniquely maps each class k to an integer 1..k for a given feature. Lastly, we can also use a method like sklearn's OneHotEncoder or pandas get_dummies, which essentially splits a given feature column with k classes into k columns, each new column with value 0 or 1.

There is a tradeoff between LabelEncoder and OneHotEncoder. The problem with LabelEncoder is that it transforms classes into integer values which are interpreted by some classifiers as continuous numeric values, which they are clearly not. For example, ['Paris', 'Tokyo', 'New York'] would be encoded as [1,2,3], with the follwoing results for some classifiers: Tokyo is interpreted as "less than" New York, averaging New York and Paris yields Tokyo, etc. OneHotEncoder, on the other hand, avoids this problem, but has problems of its own. By splitting one k-multiclass feature into k features it can drastically increase the dimensionality of the feature set and reduce classifier performance, for example if we were to binarize each value in the "occupation recode" class we are effectively adding 100 new features. Not to mention performance speed... 

We'll first try LabelEncoder. If our models perform poorly or we have time, we'll go back and try OneHot/binarized encoding.

First, let's append our dataframes and then encode. This makes sure that all of our categorical variables are encoded the same way. The problem with fitting a label encoder on our training set and then transforming the test set accordingly is that there might be values in the test set not included in the training set (e.g. "country of birth mother" == "Guam") in which case our encoder would throw an error for unencountered classes.

## Modeling

OK, we're ready for the models!

We'll start with logistic regression and use cross validation to score.

```python
# logistic regression
from sklearn import linear_model
from sklearn import cross_validation

num_folds = 10
num_instances = len(train_X)
seed = 6
kfold = cross_validation.KFold(n=num_instances, n_folds=num_folds, random_state=seed, shuffle=True)

model = sklearn.linear_model.LogisticRegression()

results = cross_validation.cross_val_score(model, train_X, train_y, cv=kfold)
print "scores:" 

print results
print '\n'
print("Accuracy: %.3f%% (+/- %.3f%%)") % (results.mean()*100.0, results.std()*100.0)
print(50 * '-')
print '\n'
#feature coefficients
model = model.fit(train_X, train_y)
print "Parameter coefficients:"
raw_coefs = model.coef_
coefs = [item for sublist in raw_coefs for item in sublist]
coefs_table = pd.DataFrame(zip(list(train_X),coefs))
coefs_table.columns = ['feature', 'coefficient value']
coefs_table
```

#### So what does a high-income individual look like? 

Based on coefficients in our model, univariate analysis, and distribution comparisons of under- and over-50K earner populations, the person most likely to make more than $50K a year is:

- Male  
- Over (roughly) 25 years old

..with additional features like race, education, and citizenship playing a less significant role. Note that in this analysis I am characterizing "significance" as a combination of predictive power as well as how much of the general population a given property extends to. For example, it is absolutely true that holding a PhD is very highly correlated with a high income, but because the population of people holding PhDs is very small relative to the entire population, this factor is less significant. It is entirely possible that under another analysis with binarized features, or under an analysis that examined only proportional distribution of a given feature against the target classes, features like "holds a PhD" would be deemed very significant.

However there are obviously a few odd things going on here: age has a relatively small coefficient even though it does plainly have a large effect on income (age under 25 and over retirement age), and wage per hour actually has a negative coefficient! This leads us to a few lackluster properties of our model affecting these results:

#### Considerations

1) 
The coefficients are going to be confused due to the fact that our categorical variables are treated like continuous ones. For something like sex which is a binary category in our survey, that poses no problem (which could be part of the reason why it has the highest coefficient score). However age, education, race, etc. are going to be treated as continuous variables and so will provide much less useful information to the classifier than they would if split out into multiple columns and binarized. 

2) 
The impact of "not in universe" responses on our data. For example, household status has the second highest coefficient value, possibly in large part because that feature contains no "not in universe" values. This feature is probably well-suited for a classification task not because it's necessarily more correlated with high income than other features in the real world, but because it contains more complete information about respondents. 

As another example, the "wage per hour" feature, should be very highly correlated with predicting whether or not someone makes more than $50K/year. In fact this feature, combined with "weeks worked per year" should be THE features that predict how much you make, but because the data was incomplete this feature was mostly "0" in our survey data and it is therefore of less consequence to our classifier. Frankly, I'm not sure why it came out as actually negative, but with more time diagnostics could perhaps tell us where the model has gone wrong.

The lesson is that the classifier balances features that do have underlying relationships with the target variable against the information-content of those features, and we should keep this in mind when attempting to use classifier results to interpret the real world.


Now, back to a confusion matrix:

```python
# confusion matrix
from sklearn import linear_model
from sklearn import cross_validation
from sklearn import metrics
model = sklearn.linear_model.LogisticRegression()
model.fit(train_X, train_y)
y_pred = cross_validation.cross_val_predict(model, train_X, train_y, cv=21)

cm = metrics.confusion_matrix(train_y, y_pred)
```

```python
# precision, recall, F-score
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(train_y, y_pred, average='binary')
```

```python
# plot confusion matrix 
def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.xticks([0,1])
    plt.yticks([0,1])
    plt.colorbar()
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

print ('Confusion matrix')    
print cm    
    
plt.figure()
plot_confusion_matrix(cm)
plt.show()
```

That's not very interpretable, so let's normalize:

```python
# normalized confusion matrix
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
print('Normalized confusion matrix')
print(cm_normalized)
plt.figure()
plot_confusion_matrix(cm_normalized, title='Normalized confusion matrix')

plt.show()
```

OK, the big insight here is that of the people who do earn more than $50,000, we're misclassifying 84.5% of them. In other words we have a very high false negative rate, which is reflected in our low recall score.

If I were to further develop this model, I would start here: looking at how and why logistic regression is misclassifying those who earn more than $50,000.

Let's try a Random Forest Classifier and see how it does.

```python
# confusion matrix
model = RandomForestClassifier(random_state=6, n_estimators=20)
model.fit(train_X, train_y)
y_pred = cross_validation.cross_val_predict(model, train_X, train_y, cv=21)

rf_cm = metrics.confusion_matrix(train_y, y_pred)
```

#### Random Forest score

Random forest classifier reduces our false negative rate, but greatly increases our false positive rate, so it's roughly a wash: the same 94% score as logistic regression. 

Again, with a more or less off-the-shelf random forest classifier we achieved a score that isn't great: just better than chance. However, it suggests that with some hyperparameter tuning (gridsearch on number of estimators, stump depth, etc.) and more careful feature selection/engineering, we could improve the performance further.
